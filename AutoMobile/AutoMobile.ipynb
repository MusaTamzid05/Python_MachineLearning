```python
import pandas as pd

from urllib.request import urlopen

df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data")
```

```python
df
```

```python
data = df.values
```

```python
import numpy as np
new_data = []
for row in data:
    if "?" in row:
        continue
    new_data.append(row)
        
```

```python
data = np.array(new_data)
print(data)
```

```python
data = pd.DataFrame(data)
data
```

```python
columns = ["symbol" , "normalize-losses" , "company" , "fuel-type" , "aspiration" , "doors" , "body-style" , 
            "drive-wheels" ,"engine-location","wheel-base","length" , "width" , "height" , "curb-width" , "engine-type" , 
           "cylinders", "engine-size","fuel-system", "bore" , "stroke" , "compression-ratio" , "hourse-power" ,
           "peak-rmp" , "city-mpg" ,"highway-mpg","price"]

data.columns = columns
data
```

```python
import numpy as np

# converting all the string to int in order to use in machine learning

class_mapping = { label:idx for idx , label in enumerate(np.unique(data["company"]))}
gas_mapping = { label:idx for idx , label in enumerate(np.unique(data["fuel-type"]))}
body_style_mapping = { label:idx for idx , label in enumerate(np.unique(data["body-style"]))}
drive_wheels_mapping = { label:idx for idx , label in enumerate(np.unique(data["drive-wheels"])) }
fuel_system_mapping = { label:idx for idx , label in enumerate(np.unique(data["fuel-system"])) }
engine_location_mapping = { label:idx for idx , label in enumerate(np.unique(data["engine-location"])) }
engine_type_mapping =  { label:idx for idx , label in enumerate(np.unique(data["engine-type"])) }
aspiration_mapping = { label:idx for idx , label in enumerate(np.unique(data["aspiration"])) }

data["company"] = data["company"].map(class_mapping)
data["fuel-type"] = data["fuel-type"].map(gas_mapping)
data["body-style"] =  data["body-style"].map(body_style_mapping)
data["drive-wheels"] = data["drive-wheels"].map(drive_wheels_mapping)
data["fuel-system"] = data["fuel-system"].map(fuel_system_mapping)
data["engine-location"] =  data["engine-location"].map(engine_location_mapping)
data["aspiration"] = data["aspiration"].map(aspiration_mapping)
data["engine-type"] = data["engine-type"].map(engine_type_mapping)

def converter(data , col):
    return [int(row) for row in data[col]]
    
data["normalize-losses"] = converter(data ,"normalize-losses" )

#data["bore"] = converter(data , "bore")
#data["stroke"] = converter(data , "stroke")
```

```python
data
```

```python

temp_dict = {"two" :2 , "three" : 3 , "four" : 4 , "six" : 6 , "five" : 5 , "eight" : 8 ,"twelve" :12 }

def str_to_int(data , col_name):
    new_row = []
    
    for row in data[col_name]:
        try:
            new_row.append(temp_dict[row])
        except TypeError:
            print("TypeError:Maybe data has been already converted")
            
    
    return new_row


data["doors"] , data["cylinders"] = str_to_int(data , "doors") , str_to_int(data , "cylinders")
```

```python
from matplotlib import pyplot as plt

plt.scatter([x for x in data["cylinders"]] , [x for x in data["hourse-power"]])
plt.xlabel("Number of Cylinders")
plt.ylabel("Hourse-Power")
plt.show()
```

```python
from sklearn.model_selection  import train_test_split



X = data[data.columns[:-1]]
y = data[data.columns[-1]]
X_train , X_test , y_train , y_test = train_test_split(X,y)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train , y_train)
```

```python
y_prediction = regression.predict(X_test)
print("score {}".format(regression.score(X_test , y_test)))
```
